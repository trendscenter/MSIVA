```python
import os
import numpy as np
import scipy.io as sio
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
root_dir = "/data/users4/xli/MSIVA/MSIVA/results/mat"
p_value_ukb = sio.loadmat(os.path.join(root_dir, "AY_group_pvalue_ukb_interaction.mat"))["p_value"]
p_value_sz = sio.loadmat(os.path.join(root_dir, "AY_group_pvalue_sz_interaction.mat"))["p_value"]
corr_ukb = sio.loadmat(os.path.join(root_dir, "AY_group_corr_ukb_interaction.mat"))["AY_corr"]
corr_sz = sio.loadmat(os.path.join(root_dir, "AY_group_corr_sz_interaction.mat"))["AY_corr"]

ms = 40

fig, axes = plt.subplots(4,5,figsize=(10,7.6))

for i in range(5):
    ax = axes[0,i]
    ln1 = ax.scatter(0, np.sum(p_value_ukb[i,1]<0.01/44318), marker='o', s=ms, color='b', label="M (+/-)")
    ax.scatter(1, np.sum(p_value_ukb[i,2]<0.01/44318), marker='o', s=ms, color='b')
    ln2 = ax.scatter(0, np.sum(p_value_ukb[i,3]<0.01/44318), marker='o', s=ms, color='r', label="F (+/-)")
    ax.scatter(1, np.sum(p_value_ukb[i,4]<0.01/44318), marker='o', s=ms, color='r')
    ax.plot([0,1], [np.sum(p_value_ukb[i,1]<0.01/44318), np.sum(p_value_ukb[i,2]<0.01/44318)], '--', color='b')
    ax.plot([0,1], [np.sum(p_value_ukb[i,3]<0.01/44318), np.sum(p_value_ukb[i,4]<0.01/44318)], '--', color='r')
    ax.set_xticks([])
    ax.set_title(f"Subspace {i+1}", fontsize=16)
    if i > 0:
        ax.set_yticks([])
    if i == 0:
        ax.set_ylabel("Number of voxels", fontsize=12)
    ax.set_xlim([-0.4,1.4])
    ax.set_ylim([12000,37000])
    
    ax2 = axes[1,i] #ax.twinx()
    ln3 = ax2.scatter(0, np.sum((p_value_ukb[i,1]<0.01/44318)&(corr_ukb[i,1]>0)), marker='^', s=ms, color='b', label="M (+)")
    ax2.scatter(1, np.sum((p_value_ukb[i,2]<0.01/44318)&(corr_ukb[i,2]>0)), marker='^', s=ms, color='b')
    ln4 = ax2.scatter(0, np.sum((p_value_ukb[i,3]<0.01/44318)&(corr_ukb[i,3]>0)), marker='^', s=ms, color='r', label="F (+)")
    ax2.scatter(1, np.sum((p_value_ukb[i,4]<0.01/44318)&(corr_ukb[i,4]>0)), marker='^', s=ms, color='r')
    ax2.plot([0,1], [np.sum((p_value_ukb[i,1]<0.01/44318)&(corr_ukb[i,1]>0)), np.sum((p_value_ukb[i,2]<0.01/44318)&(corr_ukb[i,2]>0))], '--', color='b')
    ax2.plot([0,1], [np.sum((p_value_ukb[i,3]<0.01/44318)&(corr_ukb[i,3]>0)), np.sum((p_value_ukb[i,4]<0.01/44318)&(corr_ukb[i,4]>0))], '--', color='r')
    ax2.set_xticks([])
    if i > 0:
        ax2.set_yticks([])
    if i == 0:
        ax2.set_ylabel("Number of voxels", fontsize=12)
    ax2.set_xlim([-0.4,1.4])
    ax2.set_ylim([7000,27000])

lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, bbox_to_anchor=(1.8, 0.7))

for i in range(5):
    ax = axes[2,i]
    ax.scatter(0, np.sum(p_value_sz[i,1]<0.01/44318), marker='o', s=ms, color=sns.color_palette("tab10")[0], label="HC (+/-)")
    ax.scatter(1, np.sum(p_value_sz[i,2]<0.01/44318), marker='o', s=ms, color=sns.color_palette("tab10")[0])
    ax.scatter(0, np.sum(p_value_sz[i,3]<0.01/44318), marker='o', s=ms, color=sns.color_palette("tab10")[1], label="SZ (+/-)")
    ax.scatter(1, np.sum(p_value_sz[i,4]<0.01/44318), marker='o', s=ms, color=sns.color_palette("tab10")[1])
    ax.plot([0,1], [np.sum(p_value_sz[i,1]<0.01/44318), np.sum(p_value_sz[i,2]<0.01/44318)], '--', color=sns.color_palette("tab10")[0])
    ax.plot([0,1], [np.sum(p_value_sz[i,3]<0.01/44318), np.sum(p_value_sz[i,4]<0.01/44318)], '--', color=sns.color_palette("tab10")[1])
    ax.set_xticks([])
    if i > 0:
        ax.set_yticks([])
    if i == 0:
        ax.set_ylabel("Number of voxels", fontsize=12)
    ax.set_xlim([-0.4,1.4])
    ax.set_ylim([12000,37000])
    
    ax2 = axes[3,i] #ax.twinx()
    ln3 = ax2.scatter(0, np.sum((p_value_sz[i,1]<0.01/44318)&(corr_sz[i,1]>0)), marker='^', s=ms, color=sns.color_palette("tab10")[0], label="HC (+)")
    ax2.scatter(1, np.sum((p_value_sz[i,2]<0.01/44318)&(corr_sz[i,2]>0)), marker='^', s=ms, color=sns.color_palette("tab10")[0])
    ln4 = ax2.scatter(0, np.sum((p_value_sz[i,3]<0.01/44318)&(corr_sz[i,3]>0)), marker='^', s=ms, color=sns.color_palette("tab10")[1], label="SZ (+)")
    ax2.scatter(1, np.sum((p_value_sz[i,4]<0.01/44318)&(corr_sz[i,4]>0)), marker='^', s=ms, color=sns.color_palette("tab10")[1])
    ax2.plot([0,1], [np.sum((p_value_sz[i,1]<0.01/44318)&(corr_sz[i,1]>0)), np.sum((p_value_sz[i,2]<0.01/44318)&(corr_sz[i,2]>0))], '--', color=sns.color_palette("tab10")[0])
    ax2.plot([0,1], [np.sum((p_value_sz[i,3]<0.01/44318)&(corr_sz[i,3]>0)), np.sum((p_value_sz[i,4]<0.01/44318)&(corr_sz[i,4]>0))], '--', color=sns.color_palette("tab10")[1])
    if i > 0:
        ax2.set_yticks([])
    ax2.set_xlim([-0.4,1.4])
    ax2.set_ylim([7000,27000])
    ax2.set_xticks([0,1],['Younger','Older'], fontsize=12)
    if i == 2:
        ax2.set_xlabel("Age group", fontsize=12)
    if i == 0:
        ax2.set_ylabel("Number of voxels", fontsize=12)

lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, bbox_to_anchor=(1, 0.7))

plt.tight_layout()
plt.savefig("figures/num_crossmodal_voxel.pdf")
```
